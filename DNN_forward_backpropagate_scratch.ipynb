```python
def xavier(num_neuron_inputs, num_neuron_outputs):
    temp1 =  np.sqrt(6) / np.sqrt(num_neuron_inputs+ num_neuron_outputs + 1)
    weights = stats.uniform.rvs(-temp1, 2 * temp1, (num_neuron_inputs, num_neuron_outputs))
    return weights
```

```python
grad_weights = np.zeros((len(grad_outputs), 16, 1))
def backward(inputs,grad_outputs,batch_size):
    for i in np.arange(batch_size):
        grad_weights[i,:] = np.tile(inputs[i,:], (1, 1)).T.dot(np.tile(grad_outputs[i, :], (1, 1)))
        
    return grad_weights
```

```python
def update(grad_weights,weights,update_function):
    grad_weights_average = np.mean(grad_weights, 0)
    weights = update_function(weights,grad_weights_average)
    return weights
```

```python
# 梯度下降法
def batch_gradient_descent(weights, grad_weights):  
    lr = 0.001
    weights_now = weights + lr * grad_weights
    return weights_now
```

```python
def softmaxwithloss(inputs, label):
    temp1 = np.exp(inputs)
    probability = temp1 / (np.tile(np.sum(temp1, 1), (inputs.shape[1], 1))).T
    temp3 = np.argmax(label, 1)   #纵坐标
    temp4 = [probability[i, j] for (i, j) in zip(np.arange(label.shape[0]), temp3)]
    loss = -1 * np.mean(np.log(temp4))
    return loss
def der_softmaxwithloss(inputs, label):
    temp1 = np.exp(inputs)
    temp2 = np.sum(temp1, 1)  #它得到的是一维的向量;
    probability = temp1 / (np.tile(temp2, (inputs.shape[1], 1))).T
    gradient = probability - label
    return gradient
```
